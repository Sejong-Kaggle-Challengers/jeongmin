```
import pandas as pd
import numpy as np

train = pd.read_csv("train.csv")
train = train.dropna()
train
```

```
train_x = train.loc[:, "Text"]
train_y = train.loc[:, "Label"]
train_y
```

```
test = pd.read_csv("test.csv")
test_x = test.loc[:, "Text"]
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
train_x_tfidf = tfidf.fit_transform(train_x)
print(tfidf.get_feature_names())
```

```
test_x_tfidf = tfidf.transform(test_x)
print(tfidf.get_feature_names())
```

```
from sklearn import linear_model, model_selection
clf = linear_model.RidgeClassifier()

scores = model_selection.cross_val_score(clf, train_x_tfidf, train_y , cv=5, scoring="f1")
scores
```

```
clf.fit(train_x_tfidf, train_y)
```

```
train_pred_y = clf.predict(train_x_tfidf)
```

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(train_y, train_pred_y))
```

```
test_pred_y = clf.predict(test_x_tfidf)
test_pred_y
```
